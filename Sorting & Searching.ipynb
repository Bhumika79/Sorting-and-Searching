# Sorting And Searching

```python
import tkinter
root=tkinter.Tk(className='~~~~~~~Sorting & Searching~~~~~~~')
root.geometry("600x400")
root['bg']='black'
def MenuS():
    def SortS():
        def InsertionS():
            label2.destroy()
            button3.destroy()
            button9.destroy()
            l = tkinter.Label( root, text = "Enter your list", font=('Cookie', 20), width=20, height=3, bg="pink", fg="blue")
            l.pack()
            e = tkinter.Entry( root, width=70)
            e.focus_set()
            e.pack()
            button16 = tkinter.Button(root, text='Exit', font=('Ink_Free', 10), width=20, height=3, bg="green", fg="white", activebackground='pink', command=root.destroy)
            button16.place(x=500, y=500)
            def Inst():
                button6.destroy()
                l.destroy()
                string=e.get()
                result = [int(i) for i in string.split(',')]
                aList=result
                for i in range(1, len(aList)):
                    key=aList[i]
                    j=i-1
                    while j>=0 and key<aList[j]:
                        aList[j+1]=aList[j]
                        j=j-1
                    else:
                        aList[j+1]=key
                l1 = tkinter.Label( root, text = "Your Sorted List", font=('Cookie', 20), width=45, height=3, bg="blue", fg="white" )
                l1.pack()
                l2 = tkinter.Label( root, text = [aList], font=('Cookie', 20), width=45, height=3, bg="cyan", fg="black", )
                l2.pack()
            button6 = tkinter.Button(root, text='Sort', font=('Cookie', 15), width=15, height=3, bg="beige", fg="black", activebackground='brown', command=Inst)
            button6.pack()
            button15 = tkinter.Button(root, text='Exit', font=('Ink_Free', 10), width=20, height=3, bg="green", fg="white", activebackground='pink', command=root.destroy)
            button15.place(x=500, y=500)
        def BubbleS():
            label2.destroy()
            button3.destroy()
            button9.destroy()
            l7 = tkinter.Label( root, text = "Enter your list", font=('Cookie', 20), width=20, height=5, bg="pink", fg="blue")
            l7.pack()
            e5 = tkinter.Entry( root, width=70)
            e5.focus_set()
            e5.pack()
            button14 = tkinter.Button(root, text='Exit', font=('Ink_Free', 10), width=20, height=3, bg="green", fg="white", activebackground='pink', command=root.destroy)
            button14.place(x=500, y=500)
            def BubS():
                button10.destroy()
                l7.destroy()
                stri=e5.get()
                print("String is :", stri)
                res = [int(i) for i in stri.split(',')]
                aLis=res
                n=len(aLis)
                for i in range(n):
                    for j in range(0, n-i-1):
                        if aLis[j]>aLis[j+1]:
                            aLis[j], aLis[j+1]=aLis[j+1], aLis[j]
                l3 = tkinter.Label( root, text = "Your Sorted List", font=('Cookie', 20), width=45, height=3, bg="blue", fg="white" )
                l3.pack()
                l4 = tkinter.Label( root, text = [aLis], font=('Cookie', 20), width=45, height=3, bg="cyan", fg="black", )
                l4.pack()
            button10 = tkinter.Button(root, text='Sort', font=('Cookie', 15), width=15, height=3, bg="beige", fg="black", activebackground='brown', command=BubS )
            button10.pack()
            button13 = tkinter.Button(root, text='Exit', font=('Cookie', 10), width=20, height=3, bg="green", fg="white", activebackground='pink', command=root.destroy)
            button13.place(x=500, y=500)
        label1.destroy()
        button1.destroy()
        button2.destroy()
        label2 = tkinter.Label(root, text='Which sort you want to perform:', font=('Courgette', 25), width=50, height=3, bg="black", fg="yellow")
        label2.pack()
        button3 = tkinter.Button(root, text='Insertion Sort',font=('Cookie', 20), width=20, height=3, bg="beige", fg="black", activebackground='brown', command=InsertionS)
        button3.pack()
        button9 = tkinter.Button(root, text='Bubble Sort',font=('Cookie', 20), width=20, height=3, bg="beige", fg="black", activebackground='brown', command=BubbleS)
        button9.pack()
        button12 = tkinter.Button(root, text='Exit', font=('Ink_Free', 10), width=20, height=3, bg="green", fg="white", activebackground='pink', command=root.destroy)
        button12.place(x=500, y=500)
    def SearchS():
        def linearSearch():
            label3.destroy()
            button6.destroy()
            button7.destroy()
            l5 = tkinter.Label( root, text = "Enter your list", font=('Cookie', 20), width=20, height=3, bg="pink", fg="blue")
            l5.pack()
            e1 = tkinter.Entry( root, width=70)
            e1.focus_set()
            e1.pack()
            l6 = tkinter.Label( root, text = "Enter search element", font=('Cookie', 15), width=15, height=3, bg="pink", fg="blue")
            l6.pack()
            e2 = tkinter.Entry( root, width=70)
            e2.focus_set()
            e2.pack()
            button17 = tkinter.Button(root, text='Exit', font=('Ink_Free', 10), width=20, height=3, bg="green", fg="white", activebackground='pink', command=root.destroy)
            button17.place(x=500, y=500)
            def LinearS():
                button11.destroy()
                l5.destroy()
                l6.destroy()
                st=e1.get()
                print("String is :", st)
                re = [int(i) for i in st.split(',')]
                st1=e2.get()
                print("String is:", st1)
                arr=re
                value=int(st1)
                left=0
                right=len(arr)-1
                while(left<=right):
                    if(arr[left]==value):
                        return left
                    elif(arr[right]==value):
                        return right
                    left+=1
                    right-=1
                return -1
            def LinearSe():
                a=LinearS()
                if(a==-1):
                    l8 = tkinter.Label( root, text = "Element not present", font=('Cookie', 20), width=45, height=3, bg="blue", fg="white" )
                    l8.pack()
                    print("Element not present")
                else:
                    l9 = tkinter.Label( root, text = "Element present at index" , font=('Cookie', 20), width=45, height=3, bg="blue", fg="white" )
                    l9.pack()
                    l10 = tkinter.Label( root, text = a , font=('Cookie', 20), width=45, height=3, bg="blue", fg="white" )
                    l10.pack()
                    print("Element present at index",a)
            button11 = tkinter.Button(root, text='Search', font=('Cookie', 15), width=15, height=3, bg="beige", fg="black", activebackground='brown', command=LinearSe)
            button11.pack()
            button18 = tkinter.Button(root, text='Exit', font=('Ink_Free', 10), width=20, height=3, bg="green", fg="white", activebackground='pink', command=root.destroy)
            button18.place(x=500, y=500)
        def binarySearch():
            label3.destroy()
            button6.destroy()
            button7.destroy()
            l11 = tkinter.Label( root, text = "Enter your list", font=('Cookie', 20), width=20, height=3, bg="pink", fg="blue")
            l11.pack()
            e3 = tkinter.Entry( root, width=70)
            e3.focus_set()
            e3.pack()
            l12 = tkinter.Label( root, text = "Enter search element", font=('Cookie', 20), width=20, height=3, bg="pink", fg="blue")
            l12.pack()
            e4 = tkinter.Entry( root, width=70)
            e4.focus_set()
            e4.pack()
            button19 = tkinter.Button(root, text='Exit', font=('Ink_Free', 10), width=20, height=3, bg="green", fg="white", activebackground='pink', command=root.destroy)
            button19.place(x=500, y=500)
            def BinaryS(arr1, low, high, x):
                button12.destroy()
                l11.destroy()
                l12.destroy()
                st2=e3.get()
                re1 = [int(i) for i in st2.split(',')]
                st3=e4.get()
                arr1=re1
                x=int(st3)
                low=0
                high=len(arr1)-1
                if high >= low:
                    mid = (high + low) // 2
                    if arr1[mid] == x:
                        return mid
                    elif arr1[mid] > x:
                        return BinaryS(arr1, low, mid - 1, x)
                    else:
                         return BinaryS(arr1, mid + 1, high, x)
                else:
                    return -1
            def BinarySe():
                st2=e3.get()
                re1 = [int(i) for i in st2.split(',')]
                st3=e4.get()
                arr1=re1
                x=int(st3)
                result = BinaryS(arr1, 0, len(arr1)-1, x)
                if result == -1:
                    l15 = tkinter.Label( root, text = "Element is not present in array" , font=('Cookie', 20), width=45, height=3, bg="blue", fg="white" )
                    l15.pack()
                    print("Element is not present in array")
                else:
                    l13 = tkinter.Label( root, text = "Element is present at index" , font=('Cookie', 20), width=45, height=3, bg="blue", fg="white" )
                    l13.pack()
                    l14 = tkinter.Label( root, text = result , font=('Cookie', 20), width=45, height=3, bg="blue", fg="white" )
                    l14.pack()
            button12 = tkinter.Button(root, text='Search', font=('Cookie', 15), width=15, height=3, bg="beige", fg="black", activebackground='brown', command=BinarySe)
            button12.pack()
            button20 = tkinter.Button(root, text='Exit', font=('Ink_Free', 10), width=20, height=3, bg="green", fg="white", activebackground='pink', command=root.destroy)
            button20.place(x=500, y=500)
        label1.destroy()
        button1.destroy()
        button2.destroy()
        label3 = tkinter.Label(root, text='Which search you want to perform:', font=('Courgette', 25), width=50, height=3, bg="black", fg="yellow")
        label3.pack()
        button6 = tkinter.Button(root, text='Linear Search',font=('Cookie', 20), width=25, height=3, bg="beige", fg="black", activebackground='brown', command=linearSearch)
        button6.pack()
        button7 = tkinter.Button(root, text='Binary Search',font=('Cookie', 20), width=25, height=3, bg="beige", fg="black", activebackground='brown', command=binarySearch)
        button7.pack()
        button21 = tkinter.Button(root, text='Exit', font=('Ink_Free', 10), width=20, height=3, bg="green", fg="white", activebackground='pink', command=root.destroy)
        button21.place(x=500, y=500)
    label1 = tkinter.Label(root, text='Select what you wanna do:', font=('Courgette', 25), width=25, height=3, bg="black", fg="yellow")
    label1.pack()
    button1 = tkinter.Button(root, text='Sorting', font=('Ink_Free', 12), width=25, height=3, bg="light blue", fg="green", activebackground='pink', command=SortS)
    button1.pack()
    button2 = tkinter.Button(root, text='Searching', font=('Ink_Free', 12), width=25, height=3, bg="light blue", fg="green", activebackground='pink', command=SearchS)
    button2.pack()   
MenuS()
root.state('zoomed')
root.mainloop()
```
